```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import spearmanr
sns.set(style="white", color_codes=True)
```

```python
trade = pd.read_csv('trade.csv')
```

```python
trade.head()
```

```python
g = sns.jointplot(x="k1", y="k2", data=trade)
```

```python
g = sns.jointplot(x="k1", y="k2", data=trade, kind="reg")
```

```python
g = sns.jointplot(x="k1", y="k2", data=trade, kind="hex")
```

```python
g = sns.jointplot(x="k1", y="k2", data=trade, kind="kde", space=0, color="g")
```

```python
g = sns.jointplot(x="k3", y="k4", data=trade, stat_func=spearmanr, color="m")
```

```python
g = sns.jointplot(x="k3", y="k4", data=trade, stat_func=spearmanr,
                  color="m").plot_joint(sns.kdeplot, zorder=0, n_levels=6)
```

```python
x, y = np.random.randn(2, 300)
g = sns.jointplot(x, y, kind="hex", stat_func=None).set_axis_labels("x", "y")
```

```python
g = sns.jointplot(x="k3", y="k4", data=trade, stat_func=spearmanr,
                  color="m", marginal_kws=dict(bins=15, rug=True),
                  annot_kws=dict(stat="r"), s=40, edgecolor="w", linewidth=1)
```
